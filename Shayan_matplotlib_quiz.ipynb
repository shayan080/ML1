## Question 7: Create a plot with twin axes, one for each dataset.

````python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 35]
y2 = [1, 2, 3, 4, 5]
````

````python
 
fig, ax1 = plt.subplots() 


ax1.set_xlabel('X data') 
ax1.set_ylabel('Y1 data', color = 'green') 
ax1.plot(x, y1, color = 'green') 
ax1.tick_params(axis ='y', labelcolor = 'black') 
 
ax2 = ax1.twinx() 
 

ax2.set_ylabel('Y2 data', color = 'blue') 
ax2.plot(x, y2, color = 'blue') 
ax2.tick_params(axis ='y', labelcolor = 'black') 
 
fig.suptitle('Twin Axis Plot', fontweight ="bold") 
 
plt.show() 
````

## Question 8: Plot a confusion matrix.

````python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np

y_true = [2, 0, 2, 2, 0, 1]
y_pred = [0, 0, 2, 2, 0, 2]
````

````python
conf_matrix = confusion_matrix(y_true, y_pred)

plt.figure()
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.show()
````

## Question 9: Plot a Precision-Recall curve.

````python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
import numpy as np

y_true = np.array([0, 0, 1, 1])
y_scores = np.array([0.1, 0.4, 0.35, 0.8])

precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
````

````python
plt.figure()
plt.plot(recall, precision, linestyle='-', color='blue')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')



plt.show()
````
